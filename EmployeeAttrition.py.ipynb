```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

I copied the two dataset and pasted and saved it in two different excel file and named them EmployeeWhoLeft.xlsx and ExistingEmployee.xlsx  respectively. Loading my datasets into two different variables.

```python
def leftEmployee():
    leftEmployees = pd.read_excel('Employee_who_left.xlsx')
    leftEmployees_count = len(leftEmployees["Emp ID"])
    leftCompany = []

    while leftEmployees_count != 0:
        leftCompany.append("NO")
        leftEmployees_count -= 1

    leftEmployees["leftCompany"] = leftCompany
    print(leftEmployees.head())
    return leftEmployees


dataLeft = leftEmployee()
dataLeft
```

```python
def existingEmployee():
    existingEmployees = pd.read_excel("ExistingEmployee.xlsx")
    existingEmployees_count = len(existingEmployees["Emp ID"])
    leftCompany = []
    
    while existingEmployees_count != 0:
        leftCompany.append("YES")
        existingEmployees_count -= 1
    print(existingEmployees.head)
    existingEmployees["leftCompany"] = leftCompany
    return existingEmployees
    
dataExist = existingEmployee()
dataExist
```

# JOINING THE DATASETS


```python

datasets =[ dataExist, dataLeft]
employees=pd.concat(datasets)
employees.to_excel('employees.xlsx')
employees
```

Importing the Dataset

```python
X = employees[["satisfaction_level","last_evaluation","average_montly_hours","time_spend_company","Work_accident","promotion_last_5years","dept","salary"]]
y = employees["leftCompany"]
```

Converting the string into number by encoding the categorical data and independent variable

```python
X = pd.get_dummies(X)
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

Split dataset into train and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=0)
```

Fitting Random Forest Regression to dataset 

```python
from sklearn.ensemble import RandomForestClassifier
classifier  = RandomForestClassifier(n_estimators=11, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
```

Predicting results

```python
y_pred = classifier.predict(X_test)
y_pred
```

Calculating and printing the model Accuracy and Precision 

```python
from sklearn.metrics import accuracy_score,precision_score
print(f'Accuracy:{accuracy_score(y_pred,y_test)*100}%')
print(f'Precision: {precision_score(y_test, y_pred)*100}%')
```

Visualising the results

```python
featImportances = pd.Series(classifier.feature_importances_, index=X.columns)
featImportances = feat_importances.nlargest(15)
featImportances.plot(kind='barh')
```

Reselecting the feature dataset in order to predict employee that are prone to leave

```python
X1 =employees.drop(['Emp ID','leftCompany'],axis=1)
y1 = employees["leftCompany"]
```

 Split dataset into train and test set

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1,y1,test_size=0.30,random_state=0)

EmployeesProneToLeave = pd.DataFrame(X_test1)
EmployeesProneToLeave["leftCompany"]  = y_test1
EmployeesProneToLeave["predicted"] = y_pred
EmployeesProneToLeave
```

 Predicting employees that will leave the company

```python
leaving = EmployeesProneToLeave[(EmployeesProneToLeave["predicted"] == 1) & (EmployeesProneToLeave["leftCompany"] != "YES") ]
leaving.to_excel("EmployeesProneToLeave.xlsx")
leaving
```
